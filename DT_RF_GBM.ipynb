```python
def measure(clf, x_test, y_test, target_name=None):
    y_pred = clf.predict(x_test)
    report = classification_report(y_test, y_pred, target_names=target_name)
    print("분류 보고서:\n")
    print(report)
    cm = confusion_matrix(y_test, y_pred)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=target_name)
    disp.plot(cmap=plt.cm.Blues)
    plt.show()
```

```python
Y = df[["고객이탈여부_Yes"]]
X = df.drop(['고객이탈여부_No', '고객이탈여부_Yes'], axis=1)
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size = .3, random_state=1234)
rf_base = RandomForestClassifier(random_state=1234)
dt_base = DecisionTreeClassifier(random_state=1234, max_depth=6)
gbm_base = GradientBoostingClassifier(random_state=1234)
```

```python
preds = []; cms = []; names = ["DT-baseline", "RF-baseline" , "gbm_base"]
models = [dt_base, rf_base , gbm_base]
for model in models:
    model.fit(x_train, y_train)
    pred = model.predict(x_test)
    preds.append(pred)
    cms.append(confusion_matrix(y_test, pred))
```

```python
# 결정 트리 그래프를 dot 형식으로 저장
dot_data = export_graphviz(dt_base, out_file=None, 
                           feature_names=X.columns, 
                           class_names=['class_0', 'class_1'],  # 클래스 이름 설정
                           filled=True, rounded=True, 
                           special_characters=True)

# 그래프 시각화
graph = graphviz.Source(dot_data)  
graph.render("dt_base")  # 파일로 저장 (옵션)
graph.view()  # 그래프 보기
```

```python
# 특성 중요도 시각화
fig, axes = plt.subplots(1, 3, figsize=(18, 6))
scaler = MinMaxScaler()

for i, model in enumerate(models):
    feature_importances = model.feature_importances_.reshape(-1, 1)
    scaled_importances = scaler.fit_transform(feature_importances).flatten()
    indices = np.argsort(scaled_importances)[::-1]
    
    axes[i].barh(range(X.shape[1]), scaled_importances[indices], align="center")
    axes[i].set_yticks(range(X.shape[1]))
    axes[i].set_yticklabels(X.columns[indices], fontsize=12)  # 폰트 크기 조정
    axes[i].invert_yaxis()  # 가장 중요한 특성이 위로 오도록
    axes[i].set_title(f"Feature Importances - {names[i]}", fontsize=16)  # 제목 폰트 크기 조정

plt.tight_layout()
plt.show()
```
