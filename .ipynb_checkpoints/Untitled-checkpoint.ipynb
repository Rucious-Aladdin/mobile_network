```python
def measure(clf, x_test, y_test, target_name=None):
    y_pred = clf.predict(x_test)
    report = classification_report(y_test, y_pred, target_names=target_name)
    print("분류 보고서:\n")
    print(report)
    cm = confusion_matrix(y_test, y_pred)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=target_name)
    disp.plot(cmap=plt.cm.Blues)
    plt.show()
```

```python
df = df[
    ['성별', '연령', '결혼여부', '부양자유무', '부양자수', '추천횟수', '영수증발급여부', '과금방식',
       '보안서비스', '백업서비스', '기술지원서비스', '데이터무제한', '데이터사용량', '고객이탈여부',
       '이탈가능점수', 'LTV', '만족도', '데이터사용료', '로밍사용료', '총로밍요금']
       ]
```

```python
df = pd.get_dummies(df)
```

```python
df.head()
```

```python
df.columns.tolist()
```

```python
Y = df[["고객이탈여부_Yes"]]
X = df.drop(['고객이탈여부_No', '고객이탈여부_Yes'], axis=1)
```

```python
X.columns
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size = .3, random_state=1234)
rf_base = RandomForestClassifier(random_state=1234)
dt_base = DecisionTreeClassifier(random_state=1234)
gbm_base = GradientBoostingClassifier(random_state=1234)
```

```python
preds = []; cms = []; names = ["RF-baseline", "DT-baseline", "gbm_base"]
models = [rf_base, dt_base, gbm_base]
for model in models:
    model.fit(x_train, y_train)
    pred = model.predict(x_test)
    preds.append(pred)
    cms.append(confusion_matrix(y_test, pred))
```
